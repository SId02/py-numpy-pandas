#### Let Convert rate column to float 

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

#### Type of Restarunt

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of Restarunt")
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data })
plt.plot(result,c="yellow",marker="o")
plt.xlabel("Type of Restaurant",c="grey",size=10)
plt.ylabel("Votes",c="grey" ,size=10)
```

#### Rating received by Restaurants

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("Rating received by Restaurants")
plt.show()
```
